```python
# import packages
import pandas as pd
import requests

from pandas.io.json import json_normalize
```

```python
# import data
df = pd.read_csv('ids.csv')
```

```python
df.shape
```

```python
url = 'https://itunes.apple.com/lookup?id='
x = df['publisher']
links = []
```

```python
# loop for links
for i in x: links.append(url+str(i))
```

```python
# trial json 
response = requests.get(links[0]) 
data = response.json()
results = data['results']
x = json_normalize(results)
```

```python
# creating the df the lazy way
df = pd.DataFrame(x)
df = df.drop(0)
df
```

```python
# loops
for link in links:
    response = requests.get(link) 
    data = response.json()
    results = data['results']
    x = json_normalize(results)
    df = df.append(x, ignore_index=True)
```

```python
df.tail()
```

```python
df.to_csv('apps.csv')
```
